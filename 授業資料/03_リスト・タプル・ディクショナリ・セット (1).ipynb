## 2.2 リスト

リストを作成する方法は、[  ]の中に要素をカンマで区切って並べるだけです。

````python
x = [88, 96, 90]
print(x)
print(type(x))
````

````python
y = ['math', 'science', 'english',"programming"]
print(y)
print(type(y))
````

#### リストに対して使える便利な関数

合計を計算するsum 関数

````python
scores = [88,96,90]
total = sum(scores)
print('合計点:{}'.format(total))
````

`sum`関数を使用するとリスト内のすべての要素について足し算をしてくれます。これなら、リストの中にいくつ要素があっても手間なく、足し算を計算できます。

リスト内に含まれる**要素の数を数える**len 関数

```len```は　**len**gth の略です🧐

````python
print(len(scores))
avg = total / len(scores)
print('平均点:{}'.format(avg))
````

## 2.3 ディクショナリ

python dict.png

　先ほどのリストはデータを入れると、0から始まる整数（インデックス）によって要素を管理しました。基本的に、Pythonが自動で番号を割り振ってくれるので、我々はデータを入れるだけで良いというシンプルなものでした。その一方で、「理科のテストの点数は何番目の要素だっけ？」というような、**何番目の要素が何を意味しているのか**、というのはわかりにくいです。

　そこで次に、ディクショナリについて学んでいきます。ディクショナリもリストと同様に、複数のデータをひとまとまりに集めて管理します。リストとの違いは、**キー(key)という見出し情報**を付けることができます。

　次の例のように、ディクショナリは`{}`で作成できます。そして`{}`の中に、\`キー\`:値`の形式で指定します。これによって、それぞれの値が何を表すデータかわかるようになります。

````python
scores = {
    'Math':90, 
    'History':95, 
    'English':70
    } # Mathなどのことを「キー」と呼びます。
print(scores)
print(type(scores))  # scores の型を表示
````

ディクショナリの要素を参照するためには、次のようにします。

````python
# キーを指定することで、キーに紐づく値を取得できます。
print(scores['Math']) 
````

ディクショナリを作成するときは`{}`でしたが、参照するときは`[]`を使います。

### ディクショナリ要素の追加と変更

　リストの時は要素の追加や変更ができました。ディクショナリでも、もちろんできます。ディクショナリの場合、**要素の追加**は、指定したキーがディクショナリ内の要素になければ追加されます。

````python
# Programmingというキーはないので、追加することになる
scores['Programming'] = 65 
print(scores)
````

一方、変更は既存のキーを指定して、変更したい値を代入します。

````python
# Mathの90を100に変更
scores['Programming'] = 100 
print(scores)
````

リストと同様に、要素の削除もできますが、**del文**という少し変わった構文を使います。

````python
# del文で削除したいキーを指定
del scores['Math'] 
print(scores)
````

## 2.4 タプル Tuple

　次はタプルです。リストとほぼ同じ特徴を持ちますが、「要素の追加、変更、削除ができない」という違いがあります。「中身が変更できないリスト」と理解しておいてもらって大丈夫です。リストは`[]`、ディクショナリーは`{}`で作成しましたが、タプルは`()`で作成します。

````python
weight = (50, 65, 75)
print(weight)
print(type(weight))
````

タプルの要素数と合計

````python
print(len(weight))
````

上の例のように、`()`で作成する以外はリストと同様に扱えます。ですが、次の例を見てみましょう。

````python
#要素の変更
weight[0] = 66
````

````python
#要素の追加
weight.append(45)
````

````python
#要素の削除
weight.remove(75)
````

タプルに対する、変更、追加、削除はすべてエラーになりましたね？先ほど説明したように、タプルは変更できません。実際のデータ分析などでは、あるデータに対して複数の処理を行います。その過程でデータが変更されることも多いです。タプルにすると変更されませんので、**誤って変更されたくないもの**はタプルにしておくと安全です。

## 2.5セット

コレクションの最後として、セットについて学習しましょう。セットもリストと似ていますが、次の４つの点でリストとは異なります。


1.   重複した値を格納することができない
2.   添え字やキーという概念がなく、特定の要素に対して代入・参照する方法が存在しない
3.   添え字がないため、要素は順序を持たない
4.   append関数の代わりにadd関数で要素を追加する




セットは、ディクショナリと同様に`{}`で作成できます。ただし、キーは設定しません。

````python
points = {10, 8, 9, 6}
print(points)
print(type(points))
````

````python
print(len(points))
````

````python
print(sum(points))
````

リストのように足し算や要素数を調べることはできます。一方で

````python
print(points[1])
````

上のように、セットの〇番目の要素を取得するということはできません。  
しかも

````python
points = {10, 8, 9, 6}
print(points)
````

`points = {10, 8, 9, 6}`  
と10,8,9,6という順番で入力しているにもかかわらず、表示される順番は  
`{8,9,10,6}`  
と順番も変わってしまっています。このように、セットでは順番がありません。  
それでは、セットはどのようなシーンで利用するのでしょうか？  
セットには、重複する値を格納できないという特徴があります。

Sets cannot have duplicates!

````python
points = {10, 8, 9, 6, 8}
print(points)
````

````python
points.add(10)
print(points)
````

上のように、重複した値は格納されません。この特徴を使って、学籍番号などのユニークな情報を管理するのに向いています。
