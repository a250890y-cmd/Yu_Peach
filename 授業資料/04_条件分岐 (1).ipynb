```

```

　これまでは上から順番に１行ずつ実行されていましたが、条件分岐によって実行する命令を変えたり、繰り返しによって、実行順序を変えることができます。image.png

　このように、文の実行順をコントロールするプログラムの構成を制御構造といい、上記の３つ（順次・分岐・繰り返し）が代表的なものになります。そしてオリエンテーションで話した通り、この３つを使いこなせればあらゆるプログラムを作成できます。

## 3.2 条件分岐の基本構造
　それでは、条件分岐による動作を実際に見てみましょう。

````python
# 偶数か奇数を把握するプログラム
num = int(input("好きな整数を入力してください>>"))
if num%2 == 0:
  print(f"{num}は偶数です")
else:
  print(f"{num}は奇数です")
````

````python
num = int(input("好きな整数を入力してください>>"))
if num == 7:
  print("ラッキーセブンですね!")
else:
  print("へー。")
````

上の例を実行するとわかるように、7を入れた場合と、それ以外とで返答が異なります。上記コードの流れを図で表すと以下のようになります。このような図をフローチャートといいます。
image.png

　プログラムとフローチャートを見比べるとわかるように、**if文**では

*   ifの後ろに処理が分岐する条件（**条件式**）を書く
*   条件が成立していたら、ifからelseの前までの文(**ifブロック**)が実行される
*   条件が成立していなかったら、else以降の文(**elseブロック**)が実行される

という流れで、実行されます。ifブロック、elseブロックとあるように条件式を満たした場合、満たさない場合で実行される文は複数記述できます。


````python
num = int(input("好きな整数を入力してください>>"))
if num == 7:
  print("ラッキーセブンですね!")
  print("私も7が好きです。")
else:
  print("へー。")
  print("私は7が好きです。")
````

　文を複数行実行する場合は、注意が必要です。Pythonでは、`if`ブロックや`else`ブロックの範囲を **インデント（字下げ）** で示す必要があります。`if～else`の間の`if`ブロックでは、かならず同じだけの字下げ（上の例だとスペース2文字）をする必要があります。

````python
num = int(input("好きな整数を入力してください>>"))
if num == 7:
  print("ラッキーセブンですね!")
  print("私も7が好きです。")
  print("ラッキーセブンですね!")
  print("私も7が好きです。")
  print("ラッキーセブンですね!")
  print("私も7が好きです。")
else:
  print("へー。")
  print("私は7が好きです。")
````

上のコードのように、ifブロックの1文目はスペース2文字、2文目はスペース1文字だと`IndentationError`ということでエラーが起きます。

| ---- | ---- |

### 論理演算子 Logical operators

| 論理演算子  |  意味  |
| ---- | ---- |
|  and  |  かつ  |
|  or  |  または  |
|  not  |  でなければ  |

論理演算子は次のように使います。

````python
score = int(input())
if 0 <= score and score <=100:
    print('正常な入力値です。')
else:
    print('異常な入力値です。')
````

## 3.4 分岐構文のバリエーション
　本日の最後に、if文のバリエーションについて説明します。

　if文には3つのバリエーションがあります。ここまでは、基本形のif-else構文を扱いましたが、他に

*   ifのみ構文
*   if-elif構文

があります。この２つの構文の紹介をしますので、違いを理解して、if文をマスターしましょう。

### ifのみの構文
　まずifのみの構文です。if-else構文では、条件式を満たさない場合にelseブロックが実行されますが、条件式を満たさない場合は何もしたくないこともあります。そのような場合は、ifのみの構文を使用します。

### if-elif構文
　if-else構文でも、ifのみの構文でも、ある1つの条件式を満たすかどうかで処理が分岐していました。フローチャート表すと以下のようになります。

image.png

　if-elif構文では、1つ目の条件式が成立しなかったときに、別の条件式で判定するというように、複数の条件式で処理を分岐させることができます。フローチャートで表すと以下のようになります。

image.png

if-elif構文の書き方は以下のようになります。
```
if 条件式1:
  ブロック1
elif 条件式2:
  ブロック2
else:
  ブロック3
```
上記の例では、elifは1つだけでしたが、**elifの数に制限はありません**。また、**最初に満たされた**条件式のブロックが実行されます。もしもすべての条件が満たされない場合に何もする必要がなければ、最後のelseブロックを省略できます。else:事態の記述が不要です。

````python
age = int(input("How old are you?"))
if age < 18:
  print("Wow, you are young!")
elif age >= 18 and age < 40:
  print("You look great for your age!")
else:
  print("I hope you have a retirement plan!")
````

## 1. Fortune Cookie / おみくじ

This program gives you random fortunes based on your lucky number!

````python
import random

# おみくじプログラム
luck = random.randint(1, 100)

if luck >= 90:
    print("🎉 大吉！Great fortune! Amazing things are coming your way!")
elif luck >= 70:
    print("😊 中吉！Good fortune! Today will be a good day!")
elif luck >= 40:
    print("😐 吉！Small fortune. Things are okay.")
elif luck >= 20:
    print("😟 凶！Bad luck... Be careful today.")
else:
    print("😱 大凶！Terrible luck! Maybe stay in bed today...")

print(f"Your luck score: {luck}/100")
````

## 2. Age Classifier / 年齢分類

Let's categorize people by their age!

````python
# 年齢分類プログラム
age = int(input("何歳ですか？ (How old are you?) >> "))

if age < 0:
    print("🤔 That's impossible! まだ生まれていません!")
elif age <= 2:
    print("👶 赤ちゃん Baby")
elif age <= 12:
    print("🧒 子供 Child")
elif age <= 19:
    print("🎒 ティーンエイジャー Teenager")
elif age <= 39:
    print("💼 若者 Young Adult")
elif age <= 64:
    print("👨‍💼 中年 Middle-aged")
elif age <= 100:
    print("👴 高齢者 Senior")
else:
    print("🧙‍♂️ 伝説の存在! Legendary being!")
````

## 3. Grade Calculator / 成績判定

Convert your test scores into letter grades!

````python
# 成績判定プログラム
score = int(input("テストの点数を入力してください (Enter your test score 0-100) >> "))

if score < 0 or score > 100:
    print("❌ 無効な点数です！Invalid score!")
elif score >= 90:
    print("🌟 秀 (A+) - Excellent! Outstanding work!")
elif score >= 80:
    print("😄 優 (A) - Great job!")
elif score >= 70:
    print("🙂 良 (B) - Good work!")
elif score >= 60:
    print("😐 可 (C) - You passed!")
else:
    print("😢 不可 (F) - Need to study more...")
````

## 4. Temperature Advisor / 気温アドバイザー

Get clothing advice based on the temperature!

````python
# 気温アドバイザー
temp = float(input("今日の気温は？(What's the temperature in °C?) >> "))

if temp >= 30:
    print("🥵 めちゃくちゃ暑い！Very hot! Wear shorts and t-shirt. Stay hydrated!")
elif temp >= 25:
    print("☀️ 暑い！Hot! Light clothing recommended.")
elif temp >= 20:
    print("😊 快適！Perfect! T-shirt weather.")
elif temp >= 15:
    print("🍂 涼しい！Cool. Bring a light jacket.")
elif temp >= 10:
    print("🧥 寒い！Cold! Wear a jacket or sweater.")
elif temp >= 0:
    print("🥶 かなり寒い！Very cold! Wear a coat!")
else:
    print("❄️ 凍える寒さ！Freezing! Bundle up with winter coat!")
````

## 5. Password Strength Checker / パスワード強度チェック

Check how strong your password is!

````python
# パスワード強度チェック
password = input("パスワードを入力してください (Enter a password) >> ")

length = len(password)
has_upper = any(c.isupper() for c in password)
has_lower = any(c.islower() for c in password)
has_digit = any(c.isdigit() for c in password)
has_special = any(c in "!@#$%^&*()_+-=[]{}|;:,.<>?" for c in password)

strength = 0
if length >= 8:
    strength += 1
if has_upper:
    strength += 1
if has_lower:
    strength += 1
if has_digit:
    strength += 1
if has_special:
    strength += 1

if strength == 5:
    print("💪 非常に強い！Very Strong! Great password!")
elif strength >= 3:
    print("😊 強い！Strong. Good password.")
elif strength >= 2:
    print("😐 普通 Medium. Could be better.")
else:
    print("😱 弱い！Weak! Please use a stronger password!")
````

## 6. Rock, Paper, Scissors Game / じゃんけんゲーム

Play rock, paper, scissors against the computer!

````python
import random

# じゃんけんゲーム
choices = ["グー (rock)", "チョキ (scissors)", "パー (paper)"]
computer = random.choice(choices)

print("じゃんけん！Rock, Paper, Scissors!")
print("0: グー (rock)")
print("1: チョキ (scissors)")  
print("2: パー (paper)")

player_choice = int(input("選択してください (Choose 0-2) >> "))

if player_choice < 0 or player_choice > 2:
    print("❌ 無効な選択！Invalid choice!")
else:
    player = choices[player_choice]
    print(f"\nあなた (You): {player}")
    print(f"コンピュータ (Computer): {computer}")
    
    if player == computer:
        print("🤝 引き分け！Draw!")
    elif (player == choices[0] and computer == choices[1]) or \
         (player == choices[1] and computer == choices[2]) or \
         (player == choices[2] and computer == choices[0]):
        print("🎉 あなたの勝ち！You win!")
    else:
        print("😢 あなたの負け！You lose!")
````
