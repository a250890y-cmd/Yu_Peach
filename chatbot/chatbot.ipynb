# The chatbot

This program is a simple chat bot, responding to a user's emotional status. Learn to "reverse engineer," ie, inspect the code, and understand how the application works. Then, modify the code to translate the resoponses to Japanese, making it unique to your response style.

このプログラムは、ユーザーの感情状態に応答するシンプルなチャットボットです。「リバースエンジニアリング」で、コードを見ながらアプリケーションを理解してください。その後、応答を日本語に翻訳したり、あなたの応答スタイルに合わせてコードを修正してください。

````python
# libraries
import random
from IPython.display import Image, display

# 各気分に対する応答のリストを定義
responses = {
    'happy': [
        "それは素晴らしいですね！笑顔でいてください！",
        "あなたが幸せであることがうれしいです！",
        "幸せは伝染します！広めてくださいね！",
        "その調子で！素敵な日を過ごしてください。",
        "笑顔があなたに似合っています！"
    ],
    'sad': [
        "それは残念ですね。悲しいと感じるのは普通のことです。",
        "誰かと話したいときはいつでもここにいますよ。",
        "大丈夫じゃない時もあります。ゆっくりと癒してください。",
        "悲しみは一時的なものです。少しずつ楽になるはずです。",
        "心配しないで、明日には少し良くなっているかもしれません。"
    ],
    'other': [
        "ここにいます。気持ちをもっと教えてください。",
        "もっと教えてもらえますか？",
        "あなたの気持ちを理解しようとしています。",
        "眠いのかもしれませんね。",
        "特に理由がなくても、そのままの気持ちを感じていいですよ。",
        "今の気持ちを少し言葉にしてみませんか？"
    ]
}

# 応答を返す関数を定義
def get_response(mood):
    # 応答がある場合はそのムードに対する応答をランダムに返す
    if mood in responses:
        display(Image(url="chatbot.jpg", width=200))
        # moodを表示
        print(mood)
        return random.choice(responses[mood])
    
    # それ以外の場合はデフォルトの応答を返す
    else:
        return random.choice(responses['other'])

# 会話をシミュレートするループ
while True:
    # 気分を入力
    mood = input("今日はどんな気分ですか？（happy, sad, または ESC でキャンセル）: ").strip().lower()
    
    # ESCキーで空文字列が返されるか、"esc"と入力された場合に終了
    if mood == '' or mood == 'esc' or mood == '\x1b':
        print("またいつでも話しかけてください！")
        break
    
    # リスポンスを取得して表示
    response = get_response(mood)
    print(response)

````
